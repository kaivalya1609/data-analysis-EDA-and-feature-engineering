```python
plt.figure(figsize=(12,5))
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
df_country=pd.read_excel('Country-code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
final_df.head()
```

```python
#### to check data types

final_df.dtypes
```

```python
final_df.columns
```

```python
country_name=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
### pie chart-top 3 country that uses zomato

plt.pie(country_val[:3],labels=country_name[:3],autopct='%1.2f%%')
```

###observation: zomato maximum requred are from india after that usa and united kingdom

```python
final_df.columns
```

```python
rating=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
rating
```

### observation:
when rating is between 4.5 to 4.9 ---- excellent
when the rating are between 4.0 to 4.4 ------ very good
when the rating is between 3.5 to 3.9 ----- good
when the rating is between 3.0 to 3.4---- average
when the rating is between 2.5 to 2.9------ average
when the rating is between 2.0 to 2.4 ----- poor

```python
rating.head()
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x='Aggregate rating',y='Rating count',data=rating)
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x='Aggregate rating',y='Rating count',hue='Rating color',data=rating)
```

# observation:
1. not rated count is very high
2.maximum numer of rating are between 2.5 to 3.4

```python
### count plot
sns.countplot(x='Rating color',data=rating)
```
